## Step 5 - Most suitable way to merge the dataframes here is via product_id column. The column has almost the same count on unique values, so inner join would not generate a data loss issue

```python
df_ords_prods['product_id'].nunique()
```

```python
df_prods['product_id'].nunique()
```

```python
# merge without overwriting to check the code
pd.merge(df_ords_prods,df_prods, on = 'product_id', indicator = True)
```

```python
# actual merge of the dataframes
df_ords_prods_merge = df_ords_prods.merge(df_prods, on = 'product_id', indicator = True)
```
