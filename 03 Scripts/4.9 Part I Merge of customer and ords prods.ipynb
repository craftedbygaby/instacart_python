# 03. Merging

```python
# check if key has the same distinct count in both dataframes
customers['user_id'].nunique()
```

```python
# check if key has the same distinct count in both dataframes
ords_prods['user_id'].nunique()
```

```python
# merge can be done in default inner join, as distinct count in the key column is the same in both datasets
ords_prods_all = ords_prods.merge(customers, on = 'user_id', indicator = True)
```

```python
# check the merge
ords_prods_all.head()
```

```python
ords_prods_all.shape
```

```python
# check if merge was done properly from both dataframes
ords_prods_all['_merge'].nunique()
```

```python
# dropping unecessary columns _merge
ords_prods_all = ords_prods_all.drop(columns = ['_merge'])
```

```python
ords_prods_all.shape
```

## Checks done after merge: _merge column checked and dropped, total of columns as it should, number os rows as it should. Successful merge completed
