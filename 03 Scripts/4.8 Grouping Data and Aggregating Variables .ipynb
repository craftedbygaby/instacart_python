```python
# creating logic for loyalty flag using .loc - Part II
ords_prods_merge.loc[(ords_prods_merge['max_order'] <= 40) & (ords_prods_merge['max_order'] > 10), 'loyalty_flag'] = 'Regular customer'
```

```python
# creating logic for loyalty flag using .loc - Part III
ords_prods_merge.loc[ords_prods_merge['max_order'] <= 10, 'loyalty_flag'] = 'New customer'
```

```python
# Step 6 - creating the new column "average_spending" with .transform()
ords_prods_merge['average_spending'] = ords_prods_merge.groupby(['user_id'])['prices'].transform(np.mean)
```

```python
ords_prods_merge[['user_id', 'average_spending']].head(100)
```

```python
# Step 6 - creating spending flags based on user_id and average spending - Part I
ords_prods_merge.loc[ords_prods_merge['average_spending'] < 10, 'type_of_spender'] = 'Low spender'
```

```python
# Step 6 - creating spending flags based on user_id and average spending - Part II
ords_prods_merge.loc[ords_prods_merge['average_spending'] >= 10, 'type_of_spender'] = 'High spender'
```

```python
# Step 6 - frequency check for new column type_of_spender
ords_prods_merge['type_of_spender'].value_counts(dropna = False)
```

```python
ords_prods_merge[['user_id', 'type_of_spender']].head(100)
```

```python
# Step 7 - creating the new column "median_days_prior_orders" with .transform()
ords_prods_merge['median_days_prior_orders'] = ords_prods_merge.groupby(['user_id'])['days_since_prior_order'].transform(np.median)
```

```python
ords_prods_merge[['user_id', 'median_days_prior_orders']].head(100)
```

```python
# Step 7 - creating frequent customer flag based on user_id and median_days_prior_orders - Part I
ords_prods_merge.loc[ords_prods_merge['median_days_prior_orders'] > 20, 'frequent_customer'] = 'Non-frequent customer'
```

```python
# Step 7 - creating frequent customer flag based on user_id and median_days_prior_orders - Part II
ords_prods_merge.loc[(ords_prods_merge['median_days_prior_orders'] > 10) & (ords_prods_merge['median_days_prior_orders'] <= 20), 'frequent_customer'] = 'Regular customer'
```

```python
# Step 7 - creating frequent customer flag based on user_id and median_days_prior_orders - Part III
ords_prods_merge.loc[ords_prods_merge['median_days_prior_orders'] <= 10, 'frequent_customer'] = 'Frequent customer'
```

```python
ords_prods_merge[['user_id', 'frequent_customer', 'type_of_spender']].head(100)
```

```python
# Step 7 - frequency check for new column 'frequent_customer'
ords_prods_merge['frequent_customer'].value_counts(dropna = False)
```

```python
# checking which rows have NaN - All are new customers that only did 1 order, so it is correct
ords_prods_merge[ords_prods_merge['frequent_customer'].isna()]
```
