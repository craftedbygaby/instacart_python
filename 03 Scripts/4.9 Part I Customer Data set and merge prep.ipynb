# 03. Data Wrangling

```python
# checking for mixed type data
for col in customers.columns.tolist():
  weird = (customers[[col]].map(type) != customers[[col]].iloc[0].apply(type)).any(axis = 1)
  if len (customers[weird]) > 0:
    print (col)
```

```python
# fixing mixed data type column
customers['First Name'] = customers['First Name'].astype('str')
```

```python
# checking if correction happend and check possible dtypes for downcasting
customers.info()
```

```python
# downcasting columns user_id, Age and n_dependents, income. Attention to user_id as key to be the same as the other dataset that was downcasted
customers['user_id'] = customers['user_id'].astype('int32')
customers['Age'] = customers['Age'].astype('int8')
customers['n_dependants'] = customers['n_dependants'].astype('int8')
customers['income'] = customers['income'].astype('int32')
```

```python
# checking downcasting
customers.info()
```

```python
customers.head()
```

```python
# checking the logic of STATE column to rename it properly
customers['STATE'].unique()
```

```python
# renaming columns for standarization
customers.rename(columns = {'First Name' : 'first_name'}, inplace = True)
customers.rename(columns = {'Surnam' : 'last_name'}, inplace = True)
customers.rename(columns = {'Gender' : 'gender'}, inplace = True)
customers.rename(columns = {'STATE' : 'us_state'}, inplace = True)
customers.rename(columns = {'Age' : 'age'}, inplace = True)
customers.rename(columns = {'n_dependants' : 'num_of_dependants'}, inplace = True)
customers.rename(columns = {'fam_status' : 'marital_status'}, inplace = True)
customers.rename(columns = {'income' : 'income_in_usd'}, inplace = True)
```

```python
customers.head()
```

```python
# renaming columns for standarization
customers.rename(columns = {'income_in_usd' : 'yearly_income'}, inplace = True)
```

```python
# checking if column date_joined is only one value - and therefore useless
customers['date_joined'].unique()
```

```python
# checking for names that contain 'cottage' and 'cheese' and '2' and 'milk' while excluding 'curd' to see if there was a data corruption or typos while inserting data
ords_prods_merge_clean.loc[
    ords_prods_merge_clean['product_name'].str.contains('cottage', case=False, na=False) &
    ords_prods_merge_clean['product_name'].str.contains('2', case=False, na=False) &
    ords_prods_merge_clean['product_name'].str.contains('milk', case=False, na=False) &
    ords_prods_merge_clean['product_name'].str.contains('cheese', case=False, na=False) &
    ~ords_prods_merge_clean['product_name'].str.contains('curd', case=False, na=False),
    ['product_id', 'product_name', 'prices']
    ].drop_duplicates().sort_values('product_name')
```
