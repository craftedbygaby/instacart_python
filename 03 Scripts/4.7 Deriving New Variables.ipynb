```python
# creating user defined if statement
def price_label(row):

  if row['prices'] <= 5:
    return 'Low-range product'
  elif (row['prices'] > 5) and (row['prices'] <= 15):
    return 'Mid-range product'
  elif row['prices'] > 15:
    return 'High range'
  else: return 'Not enough data'
```

##### Decided for top 5 and lower 5 as most order and fewer orders threshold (10, 11, 14, 15 and 13 as highest and 3, 4, 2, 5 and 1 as lowest). The rest is average

```python
# creating the loop
busy_hours = []

for value in df_ords_prods['order_hour_of_day']:
  if value in (10, 11, 14, 15, 13):
    busy_hours.append('Most orders')
  elif value in (3, 4, 2, 5, 1):
    busy_hours.append('Fewest orders')
  else:
    busy_hours.append('Average orders')
```

```python
# creating the new column requested
df_ords_prods['busiest_periods_of_day'] = busy_hours
```
